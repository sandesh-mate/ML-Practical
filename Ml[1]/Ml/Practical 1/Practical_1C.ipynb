```python
a=([1,2,5,7,12,4,5,6,8])
```

```python
import statistics
m=statistics.mean(a)
n=statistics.median(a)
o=statistics.mode(a)
p=statistics.stdev(a)
q=statistics.variance(a)
print("Mean is: ", m)
print("Median is: ", n)
print("Mode is: ", o)
print("Standard Deviation is: ", p)
print("Variance is: ", q)
```
